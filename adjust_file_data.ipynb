```python
import pandas as pd
import datetime
```

```python
#load dữ liệu vào data frame
df = pd.read_excel("D:\DA_Camp\data.xlsx")
df
```

```python
# Đưa dòng đầu lên làm headers
df.columns = df.iloc[0]
df = df[1:]
df
```

```python
# unpivot
# Lưu ý: đặt noOrders có nghĩa là number of Orders
df_melted = df.melt(id_vars= df.columns[0], var_name= "create_datenum", value_name= "noOrders")
df_melted
```

```python
# Tạo thêm cột ngày từ việc quy đổi 43770 đến 43869 ra 1/11/2019 đến 8/2/2020
list = df_melted["create_datenum"].unique()
dict = {}
day = datetime.date(2019, 11, 1)
tdelta = datetime.timedelta(days = 1)
for item in list:
    dict[item] = str(day)
    day = day + tdelta

list = []
for i in df_melted.index:
    list.append(dict[df_melted['create_datenum'][i]])
df_melted["create_date"] = list

df_melted
```

```python
# Bỏ cột create_datenum vì không cần thiết
df_melted = df_melted.drop(labels= 'create_datenum', axis=1)
```

```python
# Lưu file
df_melted.to_csv("D:\\DA_Camp\\adjusted_data.csv", index = False)
```
